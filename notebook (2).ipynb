## Prueba las hip√≥tesis estad√≠sticas

[Prueba la hip√≥tesis de que son diferentes los ingresos promedio procedentes de los usuarios de los planes de llamada Ultimate y Surf.]

[Elabora las hip√≥tesis nula y alternativa, escoge la prueba estad√≠stica, determina el valor alfa.]

```python
#H0= Los ingresos promedio de los usuarios del plan Ultimate y Surf son iguales.
#H1= Los ingresos promedio de los usuarios del plan Ultimate y Surf son diferentes 
from scipy import stats

from scipy import stats

# Filtrar ingresos para usuarios Ultimate y Surf en el DataFrame correcto (df)
ultimate_revenue = df[df['plan'] == 'ultimate']['total_revenue'].dropna()
surf_revenue = df[df['plan'] == 'surf']['total_revenue'].dropna()

# Prueba de hip√≥tesis H0: medias iguales vs H1: medias diferentes
# Usaremos t-test para muestras independientes (two-sample t-test)
t_stat, p_value = stats.ttest_ind(ultimate_revenue, surf_revenue, equal_var=False)

print("Estad√≠stico t:", t_stat)
print("Valor p:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hip√≥tesis nula: los ingresos promedio son diferentes.")
else:
    print("No se rechaza la hip√≥tesis nula: no hay evidencia suficiente para decir que los ingresos promedio son diferentes.")

# Descripci√≥n estad√≠stica
print("Ultimate ingresos:", ultimate_revenue.describe())
print("Surf ingresos:", surf_revenue.describe())
```
